Travel, Tourism & Hospitality - Customer Retention and Dynamic Pricing Analysis 

Week 1: Data Acquisition, Cleaning, and Feature Engineering 

Import Libraries

```python
import pandas as pd
import numpy as np
```

Load Dataset


```python
df = pd.read_excel("hotel_bookings.xls")
df.head()
```

Overview of dataset


```python
df.shape
```

```python
df.info
```

Column Standardization

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

Remove Empty Rows

```python
df = df.dropna(how='all')
```

Handle Missing Values

```python
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('unknown')
```

Handle Outliers

```python
df = df[df['average_daily_rate'] >= 0]
df = df[df['average_daily_rate'] < 500]
```

Feature Engineering

```python
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']
df['is_family'] = np.where(df['total_guests'] > 1, 1, 0)
```

Date Conversion

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

Final Check

```python
df.info()
```

Save Dataset

```python
df.to_csv("cleaned_hotel_bookings.csv", index=False)
```

### Week 1 Summary
- Removed empty rows
- Handled missing values using imputation
- Treated outliers in Average Daily Rate
- Created new features: total nights, total guests, and family indicator
- Dataset is now clean and ready for EDA
