Setup

```python
import pandas as pd

df = pd.read_csv("cleaned_hotel_bookings.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

Cancellation Distribution

```python
sns.countplot(x='cancellation_status', data=df)
plt.title("Booking Cancellation Distribution")
plt.show()
```

Lead Time vs Cancellation

```python
sns.boxplot(x='cancellation_status', y='lead_time', data=df)
plt.title("Lead Time vs Cancellation")
plt.show()
```

Average Daily Rate vs Cancellation

```python
sns.boxplot(x='cancellation_status', y='average_daily_rate', data=df)
plt.title("ADR vs Cancellation")
plt.show()
```

Deposit Type vs Cancellation

```python
sns.countplot(x='deposit_type', hue='cancellation_status', data=df)
plt.title("Deposit Type vs Cancellation")
plt.show()
```

Monthly Booking Trends (Seasonality)

```python
sns.countplot(x='arrival_date_month', data=df,
              order=df['arrival_date_month'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Monthly Booking Trends")
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Customer Segmentation

```python
sns.countplot(x='market_segment', data=df)
plt.xticks(rotation=45)
plt.title("Customer Segments")
plt.show()
```
